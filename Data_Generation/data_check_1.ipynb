```python
import nbformat
from nbconvert import PythonExporter
from IPython.core.interactiveshell import InteractiveShell


def extract_variables_from_notebook(notebook_path):
    # Load the notebook
    with open(notebook_path) as f:
        nb = nbformat.read(f, as_version=4)

    # Export the notebook to a Python code string
    exporter = PythonExporter()
    source, _ = exporter.from_notebook_node(nb)

    # Execute the code string and collect the variables
    shell = InteractiveShell.instance()
    shell.run_cell(source)

    # Get the current user namespace
    user_ns = shell.user_ns.copy()

    return user_ns
```
